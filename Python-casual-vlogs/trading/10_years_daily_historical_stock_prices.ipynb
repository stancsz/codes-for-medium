# README
This is a yfinance-based script for retrieving 10 years of stock data for quick financial analysis. This is intended to serve as a starting point for people interested in learning quant. Please do not use for trading unless you know what you're doing.

```python
# installing required libraries # copiedfrom https://stackoverflow.com/questions/44210656/how-to-check-if-a-module-is-installed-in-python-and-if-not-install-it-within-t
import sys, subprocess, pkg_resources
required = {'yfinance', 'pandas', 'pandas-datareader'}
installed = {pkg.key for pkg in pkg_resources.working_set}
missing = required - installed
if missing:
    python = sys.executable
    subprocess.check_call([python, '-m', 'pip', 'install', *missing], stdout=subprocess.DEVNULL)
```

```python
import yfinance as yf
import pandas as pd
import json
from datetime import date
from pandas_datareader import data as pdr
# Settings
## Sometime it would be convenient to visualize all rows
pd.set_option('display.max_rows', None)
## dataframe override
yf.pdr_override()
```

```python
ticker="AAPL"
today = date.today()
yft = yf.Ticker(ticker)
```
